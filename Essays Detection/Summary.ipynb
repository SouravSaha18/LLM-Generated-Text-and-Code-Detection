```python
# Install necessary libraries
!pip install matplotlib seaborn scikit-learn

# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc

# Sample data: Model performance metrics
data = {
    'Model': ['CNN-TextCNN', 'RNN-GRU', 'Hybrid LSTM-CNN', 'DistilBERT'],
    'Accuracy': [0.9738, 0.93, 0.9773, 0.9860],
    'Precision': [0.9736, 0.92, 0.9773, 0.9860],
    'Recall': [0.9738, 0.93, 0.9773, 0.9860],
    'F1-Score': [0.9736, 0.92, 0.9768, 0.9860]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the figure size for plots
plt.figure(figsize=(14, 12))

# 2. Heatmap for Performance Metrics
plt.subplot(3, 2, 2)
sns.heatmap(df.set_index('Model').T, annot=True, cmap='viridis', cbar=True)
plt.title('Heatmap of Performance Metrics')
plt.ylabel('Metrics')
plt.xlabel('Model')



# Show all plots
plt.tight_layout()
plt.show()
```
