```python
# Install necessary libraries
!pip install matplotlib seaborn scikit-learn

# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# New data: Model performance metrics
data = {
    'Model': ['CNN', 'RNN-GRU', 'Hybrid LSTM-CNN', 'DistilBERT-based Transformer'],
    'Accuracy': [0.99, 0.98, 0.98, 0.97],
    'Precision': [0.99, 0.98, 0.98, 0.97],
    'Recall': [0.99, 0.98, 0.98, 0.97],
    'F1-Score': [0.99, 0.98, 0.98, 0.97]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the figure size for plots
plt.figure(figsize=(14, 12))

# Heatmap for Performance Metrics
plt.subplot(3, 2, 1)  # Adjusting subplot position as needed
sns.heatmap(df.set_index('Model').T, annot=True, cmap='viridis', cbar=True)
plt.title('Heatmap of Performance Metrics')
plt.ylabel('Metrics')
plt.xlabel('Model')

# Show all plots
plt.tight_layout()
plt.show()
```
